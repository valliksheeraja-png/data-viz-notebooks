```python
import seaborn as sns
df=sns.load_dataset("iris")
df
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['sepal_length'], kde=True)
```

```python
sns.countplot(y='species',data=df)
```

```python
sns.barplot(x='petal_length',y='petal_width',data=df)

```

```python
sns.boxplot(x='sepal_length',y='petal_length',data=df,orient='v',palette='rainbow')
```

```python
sns.violinplot(x='sepal_length',y='petal_length',data=df,orient='v',palette='rainbow')
```
